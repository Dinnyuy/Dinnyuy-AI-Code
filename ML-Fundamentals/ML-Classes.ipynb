```python
#1 Mean, Median, Mode
import pandas
data = [12, 18, 14, 20, 16]

#calculating mean
mean = sum(data) / len(data)
print(f"Mean: {mean}")

#Calculating Median
sorted_data = sorted(data)
n = len(sorted_data)
median = (sorted_data[n // 2] + sorted_data[(n-1) //2]) / 2
print(f"Median: {median}")

#Calculating mode
from statistics import  mode
mode_value = mode(data)
print(f"Mode: {mode_value}")
```

```python
#2 Standard Deviation
import math

data = [12, 18, 14, 20, 16]
mean = sum(data)/len(data)
variance=sum((x - mean) **2 for x in data) / len(data)
std_deviation = math.sqrt(variance)

print(f"Standard Deviation: {std_deviation}")
```

```python
#3 Percentiles
import numpy as np
import matplotlib.pyplot as plt
data = [12, 14, 18, 20, 16]
p_25 = np.percentile(data, 25)
p_50 = np.percentile(data, 50)
p_75 = np.percentile(data, 75)

print(f" the first quartile is: {p_25}")
print(f" the median is: {p_50}")
print(f" the third quartile is: {p_75}")

plt.boxplot(data)
plt.title('Boxplot with Percentile')
plt.xlabel('Data')
plt.ylabel('Values')
plt.show()
```

```python
#4 Data Distribution
import matplotlib.pyplot as plt
import numpy as np

#Data distribution
data = np.random.normal(loc=0, scale=1, size=100)

plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram of Data Distribution')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
#Normal distribution
exam_scores=np.random.normal(loc=70,scale=10, size=100)
plt.hist(exam_scores, bins=30, edgecolor='black')
plt.title('Normal Distribution of exam scores in Python')
#mean=mode=median
plt.xlabel('scores')
plt.ylabel('Frequency')
plt.show()

mean_value = np.mean(data)
median_value = np.median(data)
sts_deviation = np.std(data)

print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Standard Deviation: {std_deviation}")
```
